```python
import pandas as pd

crime_data = pd.DataFrame({
    'Date': ['2023-10-27', '2023-10-27', '2023-10-28', '2023-10-28', '2023-10-29'],
    'Time': ['12:00', '18:00', '03:00', '21:00', '10:00'],
    'Location': ['India', 'Delhi', 'mumbai', 'chennai', 'bangalore'],
    'Crime Type': ['Theft', 'Assault', 'Robbery', 'Theft', 'Vandalism']
})
```

```python
import matplotlib.pyplot as plt

crime_type_counts = crime_data['Crime Type'].value_counts()
plt.bar(crime_type_counts.index, crime_type_counts.values)
plt.xlabel('Crime Type')
plt.ylabel('Frequency')
plt.title('Crime Type Frequency')
plt.show()

location_counts = crime_data['Location'].value_counts()
plt.pie(location_counts.values, labels=location_counts.index, autopct='%1.1f%%')
plt.title('Crime Location Distribution')
plt.show()
```

```python
import folium

locations = {
    'India': [20.5937, 78.9629],
    'Delhi': [28.7041, 77.1025],
    'Mumbai': [19.0760, 72.8777],
    'Chennai': [13.0827, 80.2707],
    'Bangalore': [12.9716, 77.5946]
}

map_center = locations['India']
crime_map = folium.Map(
    location=map_center,
    zoom_start=13,
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='Tiles &copy; Esri &mdash; Source: Esri, i-cubed, USDA, USGS, AEX, GeoEye, Getmapping, Aerogrid, IGN, IGP, UPR-EGP, and the GIS User Community'  # Add attribution here
)

for location, coordinates in locations.items():
    folium.Marker(location=coordinates, popup=location).add_to(crime_map)
crime_map
```
